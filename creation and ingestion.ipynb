```python
%sql
use catalog learning_spark_sql;
use schema practice_db;
update students_practice_table set marks = null, email = null;
select * from students_practice_table order by id asc;
```

```python
%sql
use catalog learning_spark_sql;
use schema practice_db;
update students_practice_table set email = case when email is null then concat(name, "@gmail.com") else "nothing" end, marks = case when marks is  null then 100 else 0 end;
select * from students_practice_table order by id asc;
--version 24
```
